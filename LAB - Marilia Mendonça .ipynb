## Verificando a popularidade das tracks dos albuns

```python
todos_os_cantos_v1['Album'] = 'Todos os cantos - Volume 1'
todos_os_cantos_v2['Album'] = 'Todos os cantos - Volume 2'
todos_os_cantos_v3['Album'] = 'Todos os cantos - Volume 3'
```

```python
def limpa_titulos(x):
    x = x.replace('\"', '')
    x = x.replace('Casa da Mãe Joana (part. Henrique & Juliano)', 'Casa da Mãe Joana - Ao Vivo')
    x = x.replace('Sabiá (part. Henrique & Juliano)', 'Sabiá - Ao Vivo')
    x = x.replace('Bebaça (part. Maiara & Maraisa)','Bebaça - Ao Vivo')
    x = x.replace('Intenção (part. Gaab)', 'Intenção - Ao Vivo')
    x = x.replace('Apaixonadinha (part. Léo Santana e Didá Banda Feminina)', 'Apaixonadinha - Ao Vivo' )
    x = x.replace('Serenata', 'Serenata - Ao Vivo')
    x = x.replace('nan', 'nnn')
    return x
```

```python
df_albuns = pd.concat([todos_os_cantos_v1, todos_os_cantos_v2, todos_os_cantos_v3])
df_albuns['Título'] = df_albuns['Título'].apply(limpa_titulos)
df_albuns = df_albuns.reset_index(drop = True)
```

```python
musicas_popularidade = []
for musica in df_albuns['Título']:
    try: 
        musicas_popularidade.append(client.search(musica)['tracks']['items'][0]['popularity'])
    except: 
        musicas_popularidade.append(0)
```

```python
musicas_popularidade
```

```python
df_albuns['Popularidade'] = pd.Series(musicas_popularidade)
df_albuns
```

##### Exportando o CSV que Irei utilizar

```python
df_albuns.to_csv("df_albuns.csv")
```
